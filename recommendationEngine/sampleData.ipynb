```python
import pandas as pd
import os
from pathlib import Path

cwd = os.getcwd()
DIR = os.path.join(cwd,"dataSalesITNozomi")
all_files = []
for filename in os.listdir(DIR):
    if filename.endswith(".xlsx"): 
        all_files.append(filename)
all_files
```

```python
all_dfs = [pd.read_excel("dataSalesITNozomi/"+x, skiprows=2, parse_dates=[1]) for x in all_files]
```

```python
import functools
for df in all_dfs:
    if "Date" not in df.columns:
        df = df.rename(columns={"Resolved Date":"Date"})
all_dfs = [df.rename(columns={"Resolved Date":"Date"}) if "Date" not in df.columns else df for df in all_dfs]
df_final = functools.reduce(lambda left,right: pd.merge(left,right,on=["Date"],how="outer"),all_dfs)
df_final
```
